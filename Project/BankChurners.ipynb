# 23. Histogram for Customer_Age

```python
import matplotlib.pyplot as plt
df['Customer_Age'].dropna().hist(bins=8, edgecolor='k')
plt.title("Histogram for Customer_Age")
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

# 24. Scatter Plot

```python
plt.figure(figsize=(8,6))
plt.scatter(df['Customer_Age'],df['Marital_Status'],alpha=0.5)
plt.xlabel('Customer_Age')
plt.ylabel('Marital_Status')
plt.title('Customer_Age vs Maritil_Status')
plt.show()
```

# 25. Box Plot

```python
attributes = ['Total_Amt_Chng_Q4_Q1', 'Total_Ct_Chng_Q4_Q1']
data = df[attributes].dropna()

plt.figure(figsize = (6,4))
data.boxplot(column = attributes)
plt.title("Boxplot of df Attributes")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```
