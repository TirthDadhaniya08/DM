# 23. Histogram for Age

```python
import matplotlib.pyplot as plt
df['age'].dropna().hist(bins=8, edgecolor='k')
plt.title("Histogram for Age")
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

# 24. Scatter Plot

```python
plt.figure(figsize=(8,6))
plt.scatter(df['chol'],df['trestbps'],alpha=0.5)
plt.xlabel('chol')
plt.ylabel('trestbps')
plt.title('chol vs trestbps')
plt.show()
```

# 25. Box Plot

```python
attributes = ['chol', 'trestbps']
data = df[attributes].dropna()

plt.figure(figsize = (6,4))
data.boxplot(column = attributes)
plt.title("Boxplot of df Attributes")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```
